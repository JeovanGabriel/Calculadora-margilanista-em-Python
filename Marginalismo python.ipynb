# Importando biblioteca

```python
#Importando a biblioteca sympy para cálculo simbólico
import sympy as sp
```

# Criação das funções e derivadas

```python
#Define a variável x (quantidade de produtos)
x = sp.Symbol('x')

#Define a função de Custo Total: C(x) = 1600 + 0.1x^2 + 25x
custo_total = 1600 + 0.1*x**2 + 25*x

#Mostra a função de custo total
print("Função de Custo Total:")
custo_total
```

```python
# Calcula a derivada da função de custo em relação a x
custo_marginal = sp.diff(custo_total, x)

#Mostra a derivada da função de custo total, que é a função de custo marginal
print("Função de Custo Marginal (Derivada do Custo Total):")
custo_marginal
```

```python
#Criamos a função de receita, assumindo um preço de venda de 60 por unidade
Receita = 60*x

#Calculamos a função de lucro, que é a receita menos o custo total
Lucro = Receita - custo_total

#Mostra a função de lucro
print("Função de Lucro:")
Lucro
```

```python
#Criamos a função de lucro marginal, que é a derivada do lucro em relação a x
Lucro_marginal = sp.diff(Lucro, x)

#Mostra a função de lucro marginal
print("Função de Lucro Marginal (Derivada do Lucro):")
Lucro_marginal
```

```python
#Criamos a função de custo médio, que é o custo total dividido pela quantidade de produtos
Custo_medio = custo_total / x

#Criamos a função de custo médio marginal, que é a derivada do custo médio em relação a x
Custo_medio_marginal = sp.diff(Custo_medio, x)

#Mostra a função de custo médio marginal
print("Função de Custo Médio Marginal:")
Custo_medio_marginal
```

# Resultados

```python
# Transforma as funções simbólicas em funções Python comuns para fazer cálculos numéricos
f_custo = sp.lambdify(x, custo_total)
f_marginal = sp.lambdify(x, custo_marginal)
f_Lucro = sp.lambdify(x, Lucro)
f_Lucro_marginal = sp.lambdify(x, Lucro_marginal)
f_custo_medio = sp.lambdify(x, Custo_medio)

# Simulando a entrada do usuário
qtd = float(input("Digite a quantidade de produtos produzidos: "))

#Mostra os resultados em relação à quantidade de produtos produzidos
print(f"\n--- Resultados para {qtd} unidades ---")
print("-----------------------------------------")
print(f"Custo Total de Produção: R$ {f_custo(qtd):.2f}")
print(f"Custo Marginal (Custo para produzir a próxima unidade): R$ {f_marginal(qtd):.2f}")
print(f"Lucro: R$ {f_Lucro(qtd):.2f}")
print(f"Lucro Marginal (Lucro para produzir a próxima unidade): R$ {f_Lucro_marginal(qtd):.2f}")
print(f"Custo Médio: R$ {f_custo_medio(qtd):.2f}")
```

```python
#Calcula a quantidade que maximiza o lucro, que é onde o lucro marginal é zero
Lucro_maximo = sp.solve(Lucro_marginal, x)
#Calcula a quantidade para sair do prejuízo e começar a lucrar, que é onde o lucro é zero
Equilibrio = sp.solve(Lucro, x)
#Calcula a quantidade que minimiza o custo médio, que é onde o custo médio marginal é zero
min_custo_medio = sp.solve(Custo_medio_marginal, x)

print(f"Quantidade que maximiza o lucro: {Lucro_maximo[0]:.0f} unidades")
print(f"Quantidade para sair do prejuízo e começar a lucrar: {Equilibrio[0]:.0f} unidades")
print(f"Quantidade que minimiza o custo médio: {min_custo_medio[1]:.0f} unidades")
```
